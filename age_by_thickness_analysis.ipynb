```python
import numpy as np
import os
import statsmodels.api as sm
age = np.loadtxt('/Users/megmcmahon/GitHub/neuropythy-for-dev/age.txt')
subjects = os.listdir('/Users/megmcmahon/GitHub/neuropythy-for-dev/data')
thickness = []
print(age)
print(subjects)
```

```python
for s in sorted(subjects):
  th = np.load(s)
  thickness.append(th)
y = np.array(thickness)

X = age
X = sm.add_constant(X)
```

```python
results = []
for i in range(163843):
  print(i)
  y = np.array(thickness)[:, i]
  model = sm.OLS(y, X)
  res = model.fit()
  results.append(res.tvalues[1])
```

```python
np.savez('results_new',results)
```
